## BONUS

The search results span multiple pages, housing a total of 631 movies in our example with each page displaying 50 movies at most. To scrape data seamlessly from all pages, you'll need to dive deep into the structure of the URLs generated with each "Next" click.

Take a close look at the following URLs:
- First page:
  ```
  https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,
  ```
- Second page:
  ```
  https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start=51&ref_=adv_nxt
  ```
- Third page:
  ```
  https://www.imdb.com/search/title/?title_type=feature&release_date=1990-01-01,1992-12-31&user_rating=7.5,&start=101&ref_=adv_nxt
  ```

You should notice a pattern. There is a `start` parameter incrementing by 50 with each page, paired with a constant `ref_` parameter holding the value "adv_nxt".

Modify your script so it's capable of iterating over all available pages to fetch data on all the 631 movies (631 is the total number of movies in the proposed example).

````python
##Create a function to update the url, and add the start parameter.

##n_pages = int(input("how many pages you want?"))
## 1---> 1 -------  1 + 50*(n_pages-1)
## 2--> 51  ------- 1 + 50*(n_pages-1) = 51
## 3--> 101 ------- 1 + 50*(n_pages-1) = 101 
````

````python
import math

def scrape_imdb(title_type, start_date, end_date, user_rating, n_movies, n_by_page):
    
    # get the number of pages
    n_pages = math.ceil(n_movies / n_by_page)
    
    url_list = []
    
    # Iterates the range based on the number of pages to get the url list
    for i in range(1,n_pages+1):
        n_pages = 1 + 50*(i-1)
        url = f"https://www.imdb.com/search/title/?title_type={title_type}&release_date={start_date},{end_date}&user_rating={user_rating},&start={n_pages}&ref_=adv_nxt"
        url_list.append(url)
        
    df = pd.DataFrame()
    
    # Iterates the list of urls to get the list of films
    for url in url_list:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')

        film_dict = {}
        index = 0
        films = soup.find_all("div", attrs={"class": "lister-item mode-advanced"})
       
        # Iterates the list of films in each page and get the information of each film
        for film in films:
            # Getting the position, name, and year
            film_header = film.find_all("h3", attrs={"class": "lister-item-header"})

            for item in film_header:
                film_position = item.find_all("span", attrs={"class": "lister-item-index unbold text-primary"})
                if film_position:
                    film_position = film_position[0].get_text().strip()
                else:
                    film_position = "n/a"
                film_name = item.find_all("a")
                if film_name:
                    film_name = film_name[0].get_text().strip()
                else:
                    film_name = "n/a"
                film_year = item.find_all("span", attrs={"class": "lister-item-year text-muted unbold"})
                if film_year:
                    film_year = film_year[0].get_text().strip()
                else:
                    film_year = "n/a"
                    
            # Getting more film data
            film_rating = film.find_all("div", attrs={"class": "inline-block ratings-imdb-rating"})
            if film_rating:
                film_rating = film_rating[0].get_text().strip()
            else:
                film_rating = "n/a"
            film_duration = film.find_all("span", attrs={"class": "runtime"})
            if film_duration:
                film_duration = film_duration[0].get_text().strip()
            else:
                film_duration = "n/a"
            film_genre = film.find_all("span", attrs={"class": "genre"})
            if film_genre:
                film_genre = film_genre[0].get_text().strip()
            else:
                film_genre = "n/a"
            film_description = film.find_all("p", attrs={"class": "text-muted"})
            if film_description:
                film_description = film_description[1].get_text().strip()
            film_director_stars = film.find_all("p")[2]  # Currently, this will return not only the director but also the actors!
            film_director = film_director_stars.find_all("a")  # Assuming the first position contains data regarding the director
            if film_director:
                film_director = film_director[0].get_text().strip()
            else:
                film_director = "n/a"
            film_actors = ",".join([actor.get_text().strip() for actor in film_director_stars.find_all("a")[1:]])  # Assuming from position 1 forward we have actor
            film_votes = film.find_all("span", attrs={"name": "nv"})
            if film_votes:
                film_votes = film_votes[0].get_text().strip()
            else:
                film_votes = "n/a"
            film_gross = film.find_all("span", attrs={"name": "nv"})
            if len(film_gross)>1:
                film_gross = film_gross[1].get_text().strip()
            else:
                film_gross = "n/a"
            
            film_dict[index] = {"ranking(movie_nr)": film_position,
                                "title": film_name,
                                "year": film_year,
                                "rating": film_rating,
                                "run_time": film_duration,
                                "genre": film_genre,
                                "description": film_description,
                                "director": film_director,
                                "stars": film_actors,
                                "votes": film_votes,
                                "gross": film_gross
                                }
            index += 1

        # Converts the dictionary to a dataframe and concats wit the one of the previous page.
        df = pd.concat([df,pd.DataFrame.from_dict(film_dict, orient="index")], axis=0)

    # Resets the index based on the updated dataframe
    df = df.reset_index(drop=True)
    return df
````
